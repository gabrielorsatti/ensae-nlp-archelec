```python
# =============================================================================
# 1. Imports
# =============================================================================
import pandas as pd
import zipfile
import os
from pathlib import Path
```

```python
# =============================================================================
# 2. Charger les métadonnées
# =============================================================================
metadata = pd.read_csv(
    "../data/archelect_search.csv",
    sep=",",
    quotechar='"',
    escapechar='\\',
    on_bad_lines='skip',
    encoding='utf-8'
)

print(f"Métadonnées : {len(metadata)} lignes, {len(metadata.columns)} colonnes")
print(f"Colonnes : {metadata.columns.tolist()}")
metadata.head()
```

```python
# =============================================================================
# 3. Explorer les métadonnées
# =============================================================================
print("Valeurs manquantes :")
print(metadata.isnull().sum())
print(f"\nÉlections disponibles : {metadata['contexte-election'].unique()}")
```

```python
# =============================================================================
# 4. Extraire les fichiers texte depuis les ZIP
# =============================================================================
data_dir = Path("../data")
text_dir = data_dir / "text_files"
text_dir.mkdir(exist_ok=True)

zip_files = list(data_dir.glob("*.zip"))
print(f"\nZIP trouvés : {[z.name for z in zip_files]}")

for zf in zip_files:
    with zipfile.ZipFile(zf, 'r') as z:
        z.extractall(text_dir)
        print(f"  {zf.name} : {len(z.namelist())} fichiers extraits")
```

```python
# =============================================================================
# 5. Charger les textes dans un DataFrame
# =============================================================================
records = []
for txt_path in text_dir.rglob("*.txt"):
    text = txt_path.read_text(encoding='utf-8', errors='replace')
    parts = txt_path.relative_to(text_dir).parts
    filename = txt_path.stem

    # Extraire année et type d'élection depuis le chemin
    year = 'unknown'
    election_type = 'unknown'
    for p in parts:
        if p in ['1981', '1988', '1993']:
            year = p
        if p in ['legislatives', 'presidentielle']:
            election_type = p

    records.append({
        'filename': filename,
        'year': year,
        'election_type': election_type,
        'text': text,
        'text_length': len(text)
    })

df_texts = pd.DataFrame(records)

print(f"\nDocuments texte chargés : {len(df_texts)}")
print(f"\nPar année :\n{df_texts['year'].value_counts()}")
print(f"\nPar type d'élection :\n{df_texts['election_type'].value_counts()}")
print(f"\nLongueur moyenne des textes par année :")
print(df_texts.groupby('year')['text_length'].mean().round(0))
```

```python
# =============================================================================
# 6. Jointure textes + métadonnées
# =============================================================================
df = df_texts.merge(metadata, left_on='filename', right_on='id', how='left')

print(f"\nJointure :")
print(f"  Textes totaux : {len(df_texts)}")
print(f"  Avec métadonnées : {df['id'].notna().sum()}")
print(f"  Sans métadonnées : {df['id'].isna().sum()}")
```

```python
# =============================================================================
# 7. Aperçu du DataFrame final
# =============================================================================
print(f"\nDataFrame final : {len(df)} lignes, {len(df.columns)} colonnes")
df[['filename', 'year', 'election_type', 'titulaire-nom', 'titulaire-soutien', 'text_length']].head(10)
```

```python
# On sauvegarde pour le deuxième notebook
df.to_pickle("../data/df_complet.pkl")
print(f"DataFrame sauvegardé : {len(df)} lignes")
```
